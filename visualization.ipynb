```python
# plot to compare baseline and hypothesis accuracies across test sets
plt.rcParams.update({'font.size': 14})

accuracy = 0.8
bright = 0.79
dark = 0.7
crop = 0.82
noisy = 0.65

baseline = {
    "original (accuracy)": accuracy,
    "bright": bright,
    "dark": dark,
    "crop": crop,
    "noisy": noisy,
}


accuracy = 0.83
bright = 0.82
dark = 0.76
crop = 0.84
noisy = 0.7

cluster0 = {
    "original (accuracy)": accuracy,
    "bright": bright,
    "dark": dark,
    "crop": crop,
    "noisy": noisy,
}


accuracy = 0.83
bright = 0.83
dark = 0.76
crop = 0.85
noisy = 0.7

cluster1 = {
    "original (accuracy)": accuracy,
    "bright": bright,
    "dark": dark,
    "crop": crop,
    "noisy": noisy,
}


accuracy = 0.68
bright = 0.67
dark = 0.56
crop = 0.71
noisy = 0.51

cluster2 = {
    "original (accuracy)": accuracy,
    "bright": bright,
    "dark": dark,
    "crop": crop,
    "noisy": noisy,
}


plt.figure(figsize=(13, 7))
plt.ylim([0, 1])
plt.ylabel("Accuracy")
plt.xlabel("Test Sets (augmentation type)")


x_pos = np.arange(len(baseline))

plt.xticks(x_pos, list(baseline.keys()))

plt.title("Accuracy score across test sets")

plt.plot(
    baseline.values(),
    marker=".",
    label="baseline"),


plt.plot(
    cluster0.values(),
    marker=".",
    label="cluster 0 (artefacts)"),

plt.plot(
    cluster1.values(),
    marker=".",
    label="cluster 1 (animals)"),

plt.plot(
    cluster2.values(),
    marker=".",
    label="cluster 2 (breeds)")
 

plt.legend(loc="lower right")
```

```python
# plot to compare baseline and hypothesis accuracies across test sets, this time with the averaged clusters score

accuracy = 0.77
bright = 0.75
dark = 0.65
crop = 0.78
noisy = 0.59

baseline = {
    "original (accuracy)": accuracy,
    "bright": bright,
    "dark": dark,
    "crop": crop,
    "noisy": noisy,
}


accuracy = (0.85 + 0.86 + 0.7) / 3
bright = (0.84 + 0.85 + 0.7) / 3
dark = (0.77 + 0.77 + 0.55) / 3
crop = (0.86 + 0.88 + 0.72) / 3
noisy = (0.72 + 0.72 + 0.5) / 3

algor = {
    "original (accuracy)": accuracy,
    "bright": bright,
    "dark": dark,
    "crop": crop,
    "noisy": noisy,
}


plt.figure(figsize=(13, 8))
plt.ylim([0.5, 1])
plt.ylabel("Accuracy")
plt.xlabel("Test Sets (augmentation type)")


x_pos = np.arange(len(baseline))

plt.xticks(x_pos, list(baseline.keys()))

plt.title("Accuracy score across test sets")

plt.plot(
    baseline.values(),
    marker=".",
    label="baseline avg={:.2f}".format(sum(baseline.values()) / len(baseline.values())),
)

plt.plot(
    algor.values(),
    marker=".",
    label="algor avg={:.2f}".format(sum(algor.values()) / len(algor.values())),
)

plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20) 
plt.legend(loc="lower right")
```

```python
# plot the accuracy per epoch of baseline

history = pd.read_csv("/mnt/c/Users/pietr/Thesis/baseline_history.csv")
print(history)


fig, ax = plt.subplots(1, figsize=(11, 6))
num_epochs = history.shape[0]

ax.plot(np.arange(0, num_epochs), history["accuracy"], label="Training accuracy")
ax.plot(np.arange(0, num_epochs), history["val_accuracy"], label="Validation accuracy")
ax.legend()

plt.tight_layout()

plt.ylim([0.5, 0.9])
plt.title("Model accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epochs")
plt.show()
```
